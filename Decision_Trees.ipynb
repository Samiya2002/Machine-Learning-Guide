# Decision Trees

Decision Tree is a Supervised learning technique that can be used for both classification and Regression problems, but mostly it is preferred for solving Classification problems. It is a tree-structured classifier, where internal nodes represent the features of a dataset, branches represent the decision rules and each leaf node represents the outcome.

In a Decision tree, there are two nodes, which are the Decision Node and Leaf Node. Decision nodes are used to make any decision and have multiple branches, whereas Leaf nodes are the output of those decisions and do not contain any further branches.The decisions or the test are performed on the basis of features of the given dataset.

It is a graphical representation for getting all the possible solutions to a problem/decision based on given conditions.

It is called a decision tree because, similar to a tree, it starts with the root node, which expands on further branches and constructs a tree-like structure.

In order to build a tree, we use the CART algorithm, which stands for Classification and Regression Tree algorithm.

A decision tree simply asks a question, and based on the answer (Yes/No), it further split the tree into subtrees.

Can consider Decision Trees as a big nested if-else statements

## Here are some important aspects to understand about decision trees:

### Splitting Criteria:
Decision trees make decisions by selecting the best attribute to split the data at each node. The goal is to find the attribute that results in the most informative splits, leading to improved class separation or reduced variance. Common splitting criteria include Gini impurity and information gain for classification problems, and reduction in variance for regression problems.

### Recursive Binary Splitting: 
Decision trees use a recursive binary splitting approach to partition the data. At each node, the selected attribute is used to split the data into two or more subsets based on different attribute values. This splitting process continues until a stopping criterion is met, such as reaching a maximum tree depth, a minimum number of samples per leaf, or when all samples belong to the same class or have the same value.

### Handling Categorical and Numerical Features: 
Decision trees can handle both categorical and numerical features. For categorical features, the tree splits the data based on the different categories. For numerical features, the tree performs binary splits by comparing the feature value with a threshold.

### Overfitting:
Decision trees are prone to overfitting, especially when they become too complex and capture noise or outliers in the training data. Overfitting occurs when the tree memorizes the training examples instead of learning general patterns. To combat overfitting, techniques like pruning, setting a minimum number of samples per leaf, or limiting the maximum tree depth can be applied.

### Ensemble Methods: 
Decision trees can be combined to form powerful ensemble models. Random Forest and Gradient Boosting are two popular ensemble methods that utilize decision trees. Random Forest builds multiple decision trees and combines their predictions, while Gradient Boosting trains decision trees sequentially, with each subsequent tree correcting the errors made by the previous trees.

### Interpretability: 
Decision trees are known for their interpretability. The decision paths from the root to the leaves can be easily understood and visualized, making it straightforward to interpret and explain the reasoning behind the predictions.

## Use of Entropy in Decision Trees?
Entropy is a concept commonly used in decision trees to measure the impurity or uncertainty of a set of data. It helps in determining the optimal attribute to split the data and build an effective decision tree. The entropy of a set is calculated using the following formula:

Entropy(S) = -Σ (p(i) * log2(p(i)))

Here, S represents the set of data, p(i) is the proportion of instances in S that belong to class i, and log2 denotes the logarithm to the base 2.

In the context of decision trees, entropy is used as a splitting criterion to determine which attribute provides the best separation of classes. The goal is to minimize the entropy in each partition or subset of the data after a split.

When deciding which attribute to use for splitting, the decision tree algorithm calculates the entropy of the target variable (class labels) before and after the split for each attribute. It then selects the attribute that results in the maximum reduction in entropy, indicating the highest information gain.

## What is Information Gain?
Information gain is a concept used in decision trees to measure the effectiveness of a particular attribute in splitting the data based on their class labels. It quantifies the reduction in entropy (or impurity) achieved by splitting the data on a specific attribute.

The information gain is calculated by comparing the entropy of the dataset before and after the split. The formula for information gain is as follows:

Information Gain = Entropy(parent) - Weighted Average Entropy(children)

Here, Entropy(parent) represents the entropy of the dataset before the split, and Weighted Average Entropy(children) represents the weighted average of the entropies of the resulting subsets or children after the split.

When deciding which attribute to use for splitting, the decision tree algorithm calculates the information gain for each attribute. The attribute with the highest information gain is chosen as the best attribute for the split. Higher information gain indicates that the attribute provides more useful and discriminative information for classifying the instances.

## What is Gini Index?
The Gini index is another impurity measure used in decision trees to evaluate the quality of a split based on class labels. It quantifies the probability of misclassifying a randomly chosen instance in a dataset.

The Gini index is calculated using the following formula:

Gini Index = 1 - Σ (p(i)^2)

Here, p(i) represents the proportion of instances in the dataset that belong to class i.

In the context of decision trees, the Gini index is used as a splitting criterion to determine the optimal attribute for dividing the data. The goal is to minimize the Gini index, as a lower value indicates a better separation of classes.

When selecting the attribute for splitting, the decision tree algorithm calculates the Gini index of the target variable (class labels) before and after the split for each attribute. It then selects the attribute that results in the minimum Gini index, implying the highest purity or separation of classes.

The Gini index ranges between 0 and 1. A Gini index of 0 indicates a perfectly pure dataset, where all instances belong to the same class. A Gini index of 1 implies maximum impurity, where the classes are evenly distributed among the instances.

## Overfitting:

Overfitting happens when a decision tree model becomes too complex and captures the noise or random fluctuations in the training data rather than the underlying patterns or relationships. It essentially memorizes the training data and performs poorly on unseen or test data. Overfitting can occur when a decision tree has too many levels (deep tree) or when it has too many branches and leaves, leading to a highly specific and complex model.
Signs of overfitting include:

High accuracy on training data but poor performance on test data.
The decision tree captures irrelevant or noisy patterns in the training data.
Excessive branching and complexity in the tree structure.
To address overfitting in decision trees, you can apply various techniques such as:

Pruning: Pruning involves simplifying the decision tree by removing unnecessary branches or nodes, which helps prevent overfitting.
Setting constraints: Limiting the depth of the tree or the minimum number of samples required to split a node can help control overfitting.
Increasing training data: Providing more diverse and representative data to the model can help it generalize better.

## Underfitting:

Underfitting occurs when a decision tree model is too simple to capture the underlying patterns in the training data. It fails to learn the relationships between the features and the target variable, resulting in poor performance on both the training and test data. Underfitting can happen when the decision tree is too shallow or when it lacks the necessary complexity to capture the patterns in the data.
Signs of underfitting include:

Low accuracy on both training and test data.
The decision tree fails to capture important patterns or relationships in the training data.
The tree structure is too simple and lacks depth.
To address underfitting in decision trees, you can try the following approaches:

Increasing the complexity: Adding more levels or branches to the decision tree can help it capture more patterns and improve performance.
Feature engineering: Enhancing the set of input features or creating new meaningful features can provide the decision tree with more information to learn from.
Using ensemble methods: Combining multiple decision trees (e.g., Random Forests or Gradient Boosting) can improve performance by reducing underfitting and leveraging the collective knowledge of the ensemble.

## Implementing Classification with Decision Trees 

```python
from sklearn.datasets import load_iris
iris=load_iris()
X=iris.data
y=iris.target
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier()
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.tree import plot_tree
from matplotlib.pylab import rcParams
```

```python
rcParams['figure.figsize'] = 80,50
plot_tree(clf)
```

## Visualizing decision trees using tools like dtreeviz can be beneficial for several reasons:

Interpretability: Decision trees can become complex, especially when they have many levels and branches. Visualizing the decision tree allows you to understand the structure and logic of the tree more intuitively. You can see the splits, branches, and leaf nodes, making it easier to interpret how the tree is making predictions.

Feature importance: Decision trees provide information about the importance of each feature in the tree's decision-making process. Visualizing the decision tree can help you identify which features have a significant impact on the predictions. You can observe which features are used for the initial splits or have a higher number of occurrences throughout the tree.

Error analysis: By visualizing the decision tree, you can identify patterns or areas where the model may be making incorrect predictions. You can examine specific paths or nodes in the tree that lead to incorrect outcomes, enabling you to analyze and address potential errors or biases in the model.

Model validation: Visualizing the decision tree can aid in model validation and debugging. You can compare the visual representation of the tree with your expectations or domain knowledge to ensure the model is behaving as desired. It helps you verify if the tree captures the intended relationships or if there are any discrepancies that need to be addressed.

Communication: Decision tree visualizations are useful for communicating your model and its predictions to stakeholders or non-technical audiences. The visual representation simplifies the understanding of the model and makes it easier to explain the decision-making process behind the predictions.
