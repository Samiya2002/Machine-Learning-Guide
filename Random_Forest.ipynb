## Random Forest

Random Forest is a popular ensemble learning method for both classification and regression tasks. It is a combination of multiple decision trees, where each tree is trained on a different subset of the data and features. The final prediction or classification is determined by aggregating the predictions of the individual trees. Random Forest's base model is always Decision Trees.

### Why is Random Forest Called so?

Random Forest is called so because it is a combination of two key components: randomness and an ensemble of decision trees.

Randomness: Randomness is incorporated into Random Forest in two main ways:

1. Random subsampling: When building each decision tree in the ensemble, Random Forest randomly selects a subset of the training data with replacement. This process is known as bootstrap sampling or bagging. By using different subsets of the data, the trees in the ensemble capture different aspects of the overall dataset, introducing randomness and reducing the risk of overfitting.

2. Feature subsampling: At each node of a decision tree, Random Forest randomly selects a subset of features to consider for splitting. This means that each tree focuses on a different subset of features. This feature subsampling helps to increase the diversity among the trees and reduces the correlation between them, making Random Forest more robust and less prone to overfitting.

Ensemble of Decision Trees: Random Forest is composed of an ensemble, or a collection, of decision trees. Each decision tree in the ensemble is constructed independently, using the random subsamples of the training data and the randomly selected features. The predictions of the individual trees are then combined, usually through majority voting (for classification) or averaging (for regression), to make the final prediction or classification.

Combining the randomness in the sampling process and the ensemble of decision trees, the method was named "Random Forest." The name reflects the idea that each tree in the ensemble is built using random subsets of the data and features, while the collective predictions of the trees form a forest of decision-making.


### Here are the key characteristics and features of Random Forest:

Ensemble of Decision Trees: Random Forest is composed of an ensemble of decision trees. Each tree is constructed independently, and they operate as a group to make predictions.

* Random Subsampling: Random Forest employs random subsampling (with replacement) of the training data to create different subsets, or "bootstrap samples." Each decision tree is trained on a different bootstrap sample, which introduces randomness and diversity into the ensemble.

* Feature Subsampling: In addition to sampling the data, Random Forest also performs feature subsampling. At each node of a decision tree, only a random subset of features is considered for splitting. This helps to reduce the correlation among the trees and makes Random Forest more robust against overfitting.

* Ensemble Prediction: For classification tasks, the predicted class is determined by majority voting among the individual trees. For regression tasks, the predictions are averaged to obtain the final output. The ensemble nature of Random Forest allows it to produce more accurate and robust predictions compared to a single decision tree.

* Handle High-Dimensional Data: Random Forest can effectively handle high-dimensional data with a large number of features. By randomly selecting a subset of features at each node, it focuses on the most informative features while reducing the impact of irrelevant or noisy features.

* Robust to Overfitting: Random Forest is less prone to overfitting compared to individual decision trees. The combination of bagging (bootstrap aggregating) and feature subsampling helps to control variance and prevents overfitting by reducing the impact of individual noisy or biased trees.

* Variable Importance: Random Forest provides a measure of variable importance, which indicates the contribution of each feature in the prediction process. This information can be used for feature selection and understanding the relationships between the features and the target variable.


### How Random Forest performs so well?

Random Forest has a wide range of applications, including classification, regression, feature selection, and anomaly detection. It is known for its robustness, versatility, and good performance on various types of datasets. Random Forest has become a popular choice in machine learning due to its ability to handle complex problems, high-dimensional data, and noisy or missing features.

Random Forest achieves low bias and low variance through the combined effect of two key techniques: bootstrap sampling (bagging) and feature subsampling. These techniques help to reduce both bias and variance, resulting in a well-balanced model. Here's how Random Forest achieves this:

* Bootstrap Sampling (Bagging):
Random Forest uses bootstrap sampling to create multiple subsets of the training data. This sampling technique involves randomly selecting data points from the original training set with replacement. By generating different bootstrap samples, each decision tree in the ensemble is exposed to slightly different subsets of the data.

The use of bootstrap sampling helps to reduce bias by allowing each tree to learn from a diverse set of training examples. It captures different aspects of the data and reduces the tendency of individual trees to overfit on specific patterns or outliers. Combining the predictions of these individually trained trees through majority voting (in classification) or averaging (in regression) helps to reduce the overall bias.

* Feature Subsampling:
In addition to sampling the data, Random Forest also performs feature subsampling. At each node of a decision tree, only a random subset of features is considered for splitting. This means that each tree focuses on a different subset of features during the learning process.

Feature subsampling helps to reduce variance by reducing the correlation among the trees. If all features were considered at each split, the trees in the ensemble would tend to make similar decisions and be highly correlated. By considering a random subset of features, each tree is forced to focus on different aspects of the data. This increases the diversity among the trees and reduces the overall variance.

By combining the benefits of bootstrap sampling and feature subsampling, Random Forest achieves a balance between bias and variance. The ensemble of decision trees captures different perspectives of the data, resulting in low bias, while the diversity among the trees reduces the overall variance.

### Random Forest Vs Bagging: 

Bagging and Random Forest are two ensemble learning techniques that use multiple models to improve predictive performance. While they share similarities, there are some key differences between the two:

Decision Tree Base Estimators:

Bagging: Bagging (Bootstrap Aggregating) can be used with any base estimator, such as decision trees, support vector machines, or neural networks. It involves training multiple instances of the same base model on different bootstrap samples of the training data.

Random Forest: Random Forest is a specific implementation of bagging that uses decision trees as base estimators. Each decision tree in the Random Forest is trained on a different bootstrap sample and a random subset of features at each split.
Feature Subsampling:

Bagging: Bagging does not involve feature subsampling. Each base estimator in bagging uses the entire set of features for training and splitting.

Random Forest: Random Forest incorporates feature subsampling. At each node of a decision tree in the Random Forest, only a random subset of features is considered for splitting. This introduces additional randomness and reduces the correlation between trees, making Random Forest more robust and effective in handling high-dimensional data.
Prediction Aggregation:

Bagging: In bagging, the predictions of all base estimators are combined by averaging (for regression) or majority voting (for classification) to obtain the final prediction.

Random Forest: Similar to bagging, Random Forest combines the predictions of all decision trees through averaging (for regression) or majority voting (for classification). However, since Random Forest uses feature subsampling, it typically exhibits lower correlation among the trees, leading to improved prediction accuracy.
Handling of Outliers:

Bagging: Bagging can be sensitive to outliers since it treats all training instances equally during bootstrap sampling. Outliers can have a strong influence on the training of individual base estimators.
Random Forest: Random Forest is more robust to outliers compared to bagging. The averaging of predictions across multiple decision trees in Random Forest helps mitigate the impact of outliers.
