# Curse Of Dimensional+ity

## The "Curse of Dimensionality" is a term used to describe the negative effects of having a large number of features or dimensions in a dataset. It refers to the fact that many machine learning algorithms and statistical models become less effective as the number of features increases, and the data becomes more sparse and harder to analyze. 

## Because of high dimensionality performance decreases and computation increases

## When working with text and image data we have high dimensionality, it is necessary to identify optimum features. In high dimensionality we face issue cause of "Sparsity". When a dataset is sparse, it means that many of the features have little or no correlation with the target variable, which can make it harder to build accurate models. In addition, sparse datasets can be more susceptible to overfitting, since there are many possible ways to fit the data to the model, and it may be difficult to identify the best one. 

## To address the curse of dimensionality, we use dimensionality reduction.  Dimensionality Reduction techniques aim to transform the high-dimensional data into a lower-dimensional space while preserving the most important information

## Dimensionality Reduction is further classified into feature selection and feature etraction.

## Feature selection involves identifying the most important features in a dataset and discarding the rest, while dimensionality reduction techniques aim to transform the high-dimensional data into a lower-dimensional space while preserving the most important information

## Feature extraction is a process of transforming raw data into a set of features that are more meaningful and informative for a specific task, such as classification or clustering. The goal of feature extraction is to reduce the dimensionality of the data and to represent it in a way that captures the essential information.

## We have few techniques in Feture Selection: Forward Selection and Backward Elimination. We also have bidirectional elimination, but these two are mostly used.

## Under Feature Extraction we have : PCA, LDA, ESNE, Wavelet Transform ICA and many more. Here we will be discussing few of them


# PCA 

## It is unsupervised machine learning algorithm. The goal of PCA is to find a set of orthogonal (uncorrelated) linear combinations of the original features that explain the maximum amount of variance in the data. These new features, called principal components, are sorted in order of importance, with the first component explaining the most variance, the second explaining the second-most, and so on.

## Benefits:
## 1. Faster Execution
## 2. Visualization

## Why variance is important for PCA?
## Variance is a key concept in PCA, as the goal of PCA is to identify the directions or components in the data that explain the most variance. In other words, PCA seeks to identify the linear combinations of the original features that maximize the amount of variation in the data.
## By focusing on the directions that explain the most variance, PCA can identify the most important patterns or structures in the data. This is important because in many high-dimensional datasets, the majority of the variance in the data is often concentrated in a small number of dimensions or directions. By identifying these important directions, PCA can help to reduce the dimensionality of the data while retaining the most important information

## STEP by STEP solution:
## 1. Find the mean and subtract from data points (not mandatory but increases performance)
## 2. Find co-variance matrix
## 3. Find eigen value and eigen vector for co-variance matrix
## 4. We get PC1,PC2.. , they are eigen values. If eigen value more then it is PC1 then PC2..
## 5. It is upto you to work on 1D, 2D...

# Implement PCA with inbuilt 

```python
from sklearn.decomposition import PCA
#Use components such that 95% of variance is retained
pca = PCA(0.95)
X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
pca.explained_variance_ratio_
```

```python
pca.n_components_
```

```python
X_pca
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train_pca, y_train)
model.score(X_test_pca, y_test)
```

```python
# Let's implement with only two components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
X_pca
```

```python
pca.explained_variance_ratio_
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)

model = LogisticRegression(max_iter=1000)
model.fit(X_train_pca, y_train)
model.score(X_test_pca, y_test)
#We get less accuancy (~60%) as using only 2 components did not retain much of the feature information. 
#However in real life you will find many cases where using 2 or few PCA components can still give you a pretty good accuracy
```

## LDA (Linear Discriminant Analysis) is a dimensionality reduction technique commonly used in machine learning and pattern recognition. It is a supervised method that seeks to find a linear combination of features that best separates different classes or categories of data.

## LDA has several advantages over other dimensionality reduction techniques such as PCA (Principal Component Analysis) and t-SNE (t-Distributed Stochastic Neighbor Embedding). It is particularly useful when the classes are well-separated and can lead to better classification accuracy than other methods. However, it assumes that the data is normally distributed and the classes have equal covariance matrices, which may not always be the case in practice.

## PCA and LDA differences:
## - In PCA we want to find the axes onto which we project our data such that maximise variance. It is unsupervised.
## -In LDA we wnat to find new axes such that class separation is maximum. It is supervised.
